# SIR Model for Epidemics



We examine the spread of a disease for which initial observations have shown that there is immunity to the virus but also a high death rate, so we decide to use the model:

$$S_{t+1} = S_{t}-\alpha S_{t}I_{t} $$
$$I_{t+1} = I_{t} + \alpha S_{t}I_{t}- \gamma I_{t} - \delta I_{t}$$
$$R_{t+1} = R_{t} + \gamma I_{t}$$
$$D_{t+1} = D_{t} + \gamma I_{t}$$

$S_{k}$: denotes the susceptible people at time k <br>
$I_{k}$: denotes the infected people at time k <br> 
$R_{k}$: denotes the recovered people at time k  <br>
$D_{k}$: denotes the people who have died from the disease at time k.


### Observations
Observations have shown that:  <br>
the percentage of people who die in a week is $\delta = 1\%$  <br>
the percentage of people who recover in a week is $\gamma = 9\%$

In a total population of $11,000$ people, all of whom could potentially get the virus, there are 1000 infected and 10000 healthy(susceptible) people at the beggining of the first week.
Also, $\alpha = \frac{\rho \gamma}{S_{1}}$. $\rho \ge 0$ with smaller $\rho$ meaning more social distancing, whereas larger $\rho$ means little to no social distancing.


```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
nweek = 100
dt = 1
beta = 1/10

gamma = 1/14

S = np.zeros(nweek)
I = np.zeros(nweek)
R = np.zeros(nweek)
t = np.arange(nweek)*dt

I[0] = 1000/11000  #percentage of infected
S[0] = 1 - I[0]
R[0] = 0

for i in range(nweek-1):
    S[i+1] = S[i] - beta * (S[i]*I[i])*dt
    I[i+1] = I[i] + (beta*S[i]*I[i]-gamma*I[i])*dt
    R[i+1] = R[i] + (gamma*I[i])*dt
    
    
fig = plt.figure(1); fig.clf()
plt.plot(t, S, 'g', label='Susceptible')
plt.plot(t, I, 'r', label='Infective')
plt.plot(t, R, 'b', label='Removed')
fig.legend(); plt.xlabel('Weeks'); plt.ylabel('Population')
```
